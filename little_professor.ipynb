```python
import random


def main():
    complexity = get_level()

    question = 0
    score = 0

    while question < 10:
        try:
            x, y =  generate_integer(complexity)
            attempt = 0

            while attempt < 3:
                z = x + y
                try:
                    answer = int(input(f'{x} + {y} = '))

                    if answer == z:
                        question += 1
                        score += 1
                        break
                    else:
                        print('EEE')
                        attempt += 1

                except ValueError:
                    pass
                except EOFError:
                    print()
                    break

                if attempt == 3:
                    question += 1
                    print(f'{x} + {y} = {z}')


        except ValueError:
            pass
        except EOFError:
            print()
            break

    print('Score: ', score)


def get_level():
    while True:
        try:
            level = int(input('Level: '))

            if level in [1, 2, 3]:
                return level
            else:
                raise ValueError
        except ValueError:
            pass
        except EOFError:
            print()
            break


def generate_integer(level):
    if level == 1:
        return random.randint(0, 9), random.randint(0, 9)
    elif level == 2:
        return random.randint(10, 99), random.randint(10, 99)
    else:
        return random.randint(100, 999), random.randint(100, 999)


if __name__ == "__main__":
    main()
```
